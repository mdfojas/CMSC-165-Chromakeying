# Background Remover

```python
import numpy as np
import cv2

#Set up the input video
video = cv2.VideoCapture('kermit.mp4')

#Get N random frames
frames_in_video = video.get(cv2.CAP_PROP_FRAME_COUNT) * np.random.uniform(size=500)

# #creating an array of frames from frames chosen above
frames = []
for frameOI in frames_in_video:
    video.set(cv2.CAP_PROP_POS_FRAMES, frameOI)
    ret, frame = video.read()
    frames.append(frame)

# #calculate the average
backgroundFrame = np.median(frames, axis=0).astype(dtype=np.uint8)

#format the background
bg = cv2.resize(backgroundFrame, (1280, 854))
cv2.imwrite("bg.jpeg",bg)

#show the background
cv2.imshow("Background",bg)
print("Done getting background")
cv2.waitKey(0)
cv2.destroyAllWindows()
```

#Chromakeying algorithm

```python
import cv2
import numpy as np
import os

video = cv2.VideoCapture("Kermit_Input.mp4")
width = int(video.get(cv2.CAP_PROP_FRAME_WIDTH ))
height = int(video.get(cv2.CAP_PROP_FRAME_HEIGHT ))
background = cv2.imread("bg.jpeg")
framespersecond= int(video.get(cv2.CAP_PROP_FPS))
totalframes = int(video.get(cv2.CAP_PROP_FRAME_COUNT ))
digits = len(str(totalframes))


if not os.path.exists("output"):
    os.mkdir("output")
    
#Set the lower and upper bound of green in HSV
l_green = np.array([30,25,25])
u_green = np.array([90,255,255])
count = 0
while True:
    #Read the frames of the input video
    ret,frame = video.read()
    
    #Check if there is a valid frame read
    if ret == False:
        print("Breaking the loop")
        break
        
    #resizing of frame and background image so they can b compatible for cross work
    frame=cv2.resize(frame, (width,height))
    background=cv2.resize(background, (width,height))
    
    #Formal start of the green screen algorithm, convert the image to HSV to start
    hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)
    
    #Creating of binary mask to detect foreground object while removing green colors
    mask =cv2.inRange(hsv, l_green, u_green)
    
    #Resulting 
    res = cv2.bitwise_and(frame, frame, mask=mask)
    foreground = frame-res
    
    #Using of simple if else statement of np: If pixel is zero, yield foreground actor, else background
    green_screen = np.where(res==0,foreground,background)
    
    #Uncommenting the 5 lines of code below will show the progress of opencv frame by frame
#     cv2.imshow("Frame",frame)
#     cv2.imshow("F",foreground)
#     cv2.imshow("Res",res)
#     cv2.imshow("Mask",mask)
    cv2.imshow("Greenscreen", green_screen)
    
    #save the frame per generated
    path = "output/img"+str(count).zfill(digits)+".jpeg"
    print(f"Saving {path}")
    count+=1
    cv2.imwrite(path,green_screen)
    
    #if esc is pressed, destroy all windows
    #a delay is used to show the images in sequence
    delay = 1000//framespersecond
    k = cv2.waitKey(delay) & 0xFF
    if k ==27:
        break
        video.release()
        cv2.destroyAllWindows()
video.release()
cv2.destroyAllWindows()
print("End algorithm")
```
